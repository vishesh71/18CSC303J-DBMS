```python
import matplotlib.pyplot as plt
import numpy as np

mean = 0
variance = 1

#Given sample sizes
N_list=[250, 1000, 2000, 4000]

#Given gaussian distribution function
def get_pdf(x):
    return 1/np.sqrt(2*np.pi)*np.exp(-x**2/2)
```

```python
# A.

'''
According to the graph below, the gaussian pdf fit on each histogram is directly proportional to the number of samples
This is due to Central Limit Theorem and Law of Large Numbers.
'''

fig,axis=plt.subplots(len(N_list),1,figsize=(8, 12))

for i,N in enumerate(N_list):
    sample_val=np.random.randn(N)

    axis[i].hist(sample_val, bins=30, density=True, color="green")

    x_val=np.linspace(-3, 3, 100)

    #Calculating and plotting the Gaussian PDF values
    pdf_val=get_pdf(x_val)
    axis[i].plot(x_val, pdf_val, color="red")
    axis[i].set_title(f'Value of N={N}')

plt.show()
```

```python
# C.
import matplotlib.pyplot as plt
import numpy as np

mean = 0
variance = 1

N_list = [250, 1000, 2000, 4000]

# Defining the function for the true pdf of the Gaussian distribution
def get_pdf(x):
    return np.sqrt(1 / (2 * np.pi)) * np.exp(-0.5 * x**2)

fig, axis = plt.subplots(len(N_list), 1, figsize=(8, 12))

for i, N in enumerate(N_list):
    sample_val = []

    for j in range(N):
      if j == 0:
        sample_val.append(np.random.randn())
        continue

      # Rolling a dice
      dice_roll = np.random.randint(1, 7)

      # adding a new random value to samples if 1
      if dice_roll == 1:
          sample_val.append(np.random.randn())

      # adding a value from previous samples
      else:
          sample_val.append(np.random.choice(sample_val))

    axis[i].hist(sample_val, bins=30, density=True, color="green")

    # Defining x values and plotting the Gaussian pdf values
    x_val = np.linspace(-3, 3, 100)
    pdf_val = get_pdf(x_val)
    axis[i].plot(x_val, pdf_val, color = "red")
    axis[i].set_title(f'value of N = {N}')

plt.tight_layout()
plt.show()
```
